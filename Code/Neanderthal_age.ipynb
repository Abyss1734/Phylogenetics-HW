```python
from Bio import AlignIO

# Задаем путь к файлу с выравниванием
file_path = r"C:\Users\rusla\OneDrive\Рабочий стол\aligned_sequences (Human + Neanderthal + Denisovan).fasta"

# Загружаем выровненные последовательности
alignment = AlignIO.read(file_path, "fasta")

# Список индексов, которые нужно исключить из расчетов
exclude_ids = ["KX198084.1", "KX198088.1", "KX198086.1", "KX198085.1", 
               "FN673705.1", "FR695060.1", "KT780370.1"]

# Ищем последовательность последнего Неандертальца по ID
mitochondrial_eve = None
for record in alignment:
    if "KX198087.1" in record.id:
        mitochondrial_eve = record
        break

if mitochondrial_eve is None:
    raise ValueError("Не удалось найти последовательность для последнего Неандертальца")

# Функция для подсчета точечных мутаций между двумя последовательностями
def count_mutations(seq1, seq2):
    mutations = 0
    for i in range(len(seq1)):
        if seq1[i] != seq2[i] and seq1[i] != "-" and seq2[i] != "-":
            mutations += 1
    return mutations

# Подсчет мутаций для каждой последовательности по сравнению с последним Неандертальцем
mutations = []
for record in alignment:
    if record.id != mitochondrial_eve.id and record.id not in exclude_ids:  # Исключаем записи по ID
        num_mutations = count_mutations(mitochondrial_eve.seq, record.seq)
        mutations.append(num_mutations)
        print(f"{record.id}: {num_mutations} мутаций")

# Находим максимальное количество мутаций
if mutations:  # Проверяем, есть ли мутации
    max_mutations = max(mutations)
    print(f"Максимальное количество мутаций: {max_mutations}")

# Используем скорость мутаций для оценки возраста
mutation_rate = 4000  # лет на одну мутацию
estimated_age = max_mutations * mutation_rate / 2
print(f"Оценочный возраст последнего Неандертальца: {estimated_age} лет")
```
